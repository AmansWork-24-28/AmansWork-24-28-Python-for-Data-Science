SET 

Set is a collection of non-repetitive elements.

✅ Stores unique elements only.
✅ Duplicate values are automatically removed.
✅ Unordered (do not rely on element positions).
❌ No indexing (s[0] is invalid).
✅ Mutable (can add/remove elements).
❌ Elements must be immutable (e.g., numbers, strings, tuples).
❌ Lists and dictionaries cannot be elements of a set because they are mutable.

```python
s = {10, 20, 30, 40, 50}
print(s)
```

SET METHODS

```python
#1. Find the length of the set
print(len(s))
print()

#2. Add a single element
s.add(60)
print(s)
print()

#3. Add multiple elements
s.update([70, 80, 90])
print(s)
print()

#4. Remove an element (gives error if not present)
s.remove(20)
print(s)
print()

#5. Remove an element (no error if not present)
s.discard(100)
print(s)
print()

#6. Remove and return a random element
print(s.pop())
print(s)
print()

#7. Copy the set
s3 = s.copy()
print(s3)
print()

#9. Clear the set
temp = s.copy()
temp.clear()
print(temp)
print()

#10. Union of two sets
print(s.union(s2))
print()

#11. Intersection of two sets
print(s.intersection(s2))
print()

#12. Difference of two sets
print(s.difference(s2))
print()

#13. Symmetric Difference
print(s.symmetric_difference(s2))
print()

#14. Update with union
s4 = s.copy()
s4.update(s2)
print(s4)
print()

#15. Intersection Update
s5 = s.copy()
s5.intersection_update(s2)
print(s5)
print()

#16. Difference Update
s6 = s.copy()
s6.difference_update(s2)
print(s6)
print()

#17. Symmetric Difference Update
s7 = s.copy()
s7.symmetric_difference_update(s2)
print(s7)
print()

#18. Check subset
print({30, 40}.issubset(s))
print()

#19. Check superset
print(s.issuperset({30, 40}))
print()

#20. Check disjoint
print(s.isdisjoint({100, 200}))
print()

#21. Check if an element exists
print(30 in s)
print()

#22. Check if an element does not exist
print(200 not in s)
print()

#23. Iterate through the set
for i in s:
    print(i)
print()

#24. Convert list to set
l = [1, 2, 2, 3, 3, 4]
print(set(l))
print()

#25. Convert set to list
print(list(s))
print()

#26. Create an empty set
empty = set()
print(empty)
print()
```
