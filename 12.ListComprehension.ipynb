# List Comprehension

List comprehension is a concise and efficient way to create lists in Python. It allows
you to generate lists in a single line of code, making your code more readable and
Pythonic.

# When to Use List Comprehensions?
You need to create a list in a single line  
The logic is simple and readable  
You want to improve performance (faster than loops)  
Avoid when: - The logic is too complex (use a standard loop instead for clarity)
  

#  1.Basic syntax

 [expression for item in iterable]  
 expresiion the operation that performs on each item  
 item - represnts each element  
 itreable - data structure(list,range)  

```python
cubes = [x**3 for x in range(1,6)]
print(cubes)
```

# 2. Using if Condition  



```python
evens = [x for x in range(1,10) if x%2==0]
print(evens)
```

# 3. Using if-else Condition

```python
nums = ["Even" if x%2==0 else "Odd" for x in range(0,10)]
print(nums)
```

# 4. Nested Loops

```python
nested = [(x,y) for x in range(2) for y in range(3)] # matrix
print(nested)
```

# 5. List Comprehension with Functions

```python
words = ["Hi","Aman","Gotad"]
upper = [w.upper() for w in words]
print(upper)

words = ["Hi","Aman","Gotad"]
lower = [w.lower() for w in words]
print(lower)
```

# 6. List Comprehension with Nested List Comprehension

```python
matrix = [[1,2],[2,5],[1,4]]
flat = [n for row in matrix for n in row] # row represnts the first element and n reprents the number in row
print(flat)
```

# 7. List Comprehension with Set and Dictionary Comprehensions

```python
# Set

list = [1,2,2,5,3,3,6,5,4,1,2]
sets = {x for x in list}
print(sets)
```

```python
# Dictionaries

dict = {x:x*2 for x in range(4)}
print(dict)
```
