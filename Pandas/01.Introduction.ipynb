# Pandas 

Pandas is a powerful, open-source Python library used for data manipulation, cleaning, and analysis.  

It provides two main data structures:    

1.Series: A one-dimensional labeled array  
2.DataFrame: A two-dimensional labeled table (like an Excel sheet or SQL table)  




```python
import pandas as pd
```

# Series

```python
s1 = pd.Series([10,15,20,25,30])
print(s1)
print()
s2 = pd.Series(["A","B","C","D"])
print(s2)

print("\n Custom index")
```

# DATA FRAME


```python
data = {
    "Name" : ["Aman"],
    "Class" : ["CE2"],
    "Div" : ["Two"]
}
df = pd.DataFrame(data)
df
```

```python
# row and column labels
df.index
df.columns

df.index = ['a']
```

# Creating DataFrames

```python
# list inside list

data = [
["Alice", 25],
["Bob", 30],
["Charlie", 35]
]

df = pd.DataFrame(data,columns=["Name","age"])
df
```

```python
# from dictionaries

data = {
"Name": ["Alice", "Bob", "Charlie"],
"Age": [25, 30, 35]
}
df = pd.DataFrame(data)
```

```python
# using numpy array
import numpy as np

arr = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(arr, columns=["A", "B"])
df
```
