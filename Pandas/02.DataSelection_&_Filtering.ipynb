# Data Selection & Filtering

```python
# Selecting Rows & Columns

df["Name"] # single column (series)
df[["Name","Age"]] # data frame
```

```python
# Selecting Rows by Index

df.loc[2] # First row (by label)
df.iloc[2] # First row (by position)

print(df.loc[0, "Name"])  # Value at row 0, column 'Name'
print(df.iloc[0, 1])  # Value at row 0, column at index 1

print(df.loc[0:2, ["Name", "Age"]]) # Rows 0 to 2, selected columns
print()
print(df.iloc[0:2, 0:2] )# Rows and cols by index position

# faster way - similar to loc just no slicing

df.at[0, "Name"] # Fast label-based access
df.iat[0, 1]  # Fast position based access
```

# Filtering with Conditions

```python
df[df["Age"]>20]
```

```python
# syntax - df[(df[condn])&(df[condn])]

df[(df["Name"]=="Bob") & (df["Age"]>20)] # using and & or |
```

```python
# Querying with .query()


df.query("Age > 20 and Name=='Ethan'")
```
