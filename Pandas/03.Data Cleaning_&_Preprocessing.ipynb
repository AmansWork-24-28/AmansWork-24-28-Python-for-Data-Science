```python
import pandas as pd

df = pd.read_csv("sample.csv")
df.shape
```

# Handling Missing Values

```python
# Check for Missing Data  

df.isnull()
```

```python
df.isnull().sum() # total null values in column
```

```python
# Drop Missing Data

df.dropna()
df.dropna(axis=1)
df.shape
```

```python
# Fill Missing Data

df.fillna(0) # fill the nans with 0
```

```python
df["Age"].fillna(df["Age"].mean())
# df["Age"].fillna(value)
```

```python
df.ffill() # give forward value to nan means above box
```

```python
df.bfill()  # give backwardward value to nan means below box
```

# Detecting & Removing Duplicates

```python
df.duplicated() # gives that is there duplicate value or not
```

```python
df.drop_duplicates()
```

```python
df.duplicated(subset=["Name", "Age"])
```
