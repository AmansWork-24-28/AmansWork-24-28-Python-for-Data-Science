```python
import json

data = {"name": "Aman", "age": 25, "city": "New York"}
```

# Converting Python Objects to JSON (Serialization)

functions used -
1.dumps
2.dump

python obj to json(str)

```python
# json.dumps() – Convert Python object to JSON string

json_string = json.dumps(data)
print(json_string)
print(type(json_string))
```

```python
# json.dump() – Write JSON data to a file

with open("sample.json","w") as file:
    json.dump(data,file) 
```

# Converting JSON to Python Objects (Deserialization)

functions used -
1.loads
2.load

json to python obj


```python
json_data = '{"name": "Alice", "age": 25, "city": "New York"}'
print(type(json_data))

py_obj = json.loads(json_data)
print(py_obj)

print(type(py_obj))
```

```python
with open("sample.json", "r") as file:
 python_data = json.load(file)
print(python_data)
```

# Formatting JSON Output


```python
formatted_json = json.dumps(data, indent=4)
print(formatted_json)
print(type(formatted_json))
```
