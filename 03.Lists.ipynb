FUNCTIONS OF LIST -

```python
# Initialize a new list with different numbers
l1 = [14, 3, 9, 22, 5, 11]
print("Original list:", l1)

# 1. l1.sort(): sorts the list in ascending order
l1.sort()
print("After sort():  ", l1)

# 2. l1.reverse(): reverses the current order of the list
l1.reverse()
print("After reverse():", l1)

# 3. l1.append(8): adds 8 at the very end of the list
l1.append(8)
print("After append(8):", l1)

# --- Resetting list to its original state for the remaining operations ---
l1 = [14, 3, 9, 22, 5, 11]

# 4. l1.insert(3, 8): inserts the number 8 at index 3
l1.insert(3, 8)
print("After insert(3, 8):", l1)

# 5. l1.pop(2): deletes the element at index 2 and returns it
popped_value = l1.pop(2)
print("After pop(2):      ", l1, f" (Popped value: {popped_value})")

# 6. l1.remove(22): searches for and removes the number 22
l1.remove(22)
print("After remove(22):  ", l1)
```
