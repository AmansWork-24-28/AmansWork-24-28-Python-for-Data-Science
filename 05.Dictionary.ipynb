Dictionary is a collection of keys-value pairs.

PROPERTIES OF PYTHON DICTIONARIES
1. It is unordered.
2. It is mutable( changable).
3. It is indexed.
4. Cannot contain duplicate keys -
   keys should be unique but the values can be duplicate
   eg.
      marks = {
    "Math": 95,
    "Physics": 90,
    "Chemistry": 95
    }
   
+ Empty dictionary is shown as
  my_dict = {}


```python
d = {
    "key" : "value",
    "Name" : "Aman",
    "PRN" : "B24CE1064",
    "Course" : "BTech"
}

d
```

```python
# Accesing the value using their specific key

print(d["Name"])  #Aman
print(d["PRN"])   #B24CE1064
```

DICTIONARY METHODS ->>


```python
print()

#1. Print all the key and value pair in dictionary
print(d.items())
print()

#2. Print all keys
print(d.keys())
print()

#3. Update value of key
d.update({"PRN": "B24CE1640"})
print(d)
print()

#4. To get the value of specified key
print(d.get("Name"))
print()

#5. Print all values
print(d.values())
print()

#6. Access value using key
print(d["PRN"])
print()

#7. Add a new key-value pair
d["Branch"] = "Computer Engineering"
print(d)
print()

#8. Remove a key using pop()
d.pop("PRN")
print(d)
print()

#9. Remove the last inserted key-value pair
d.popitem()
print(d)
print()

#10. Check if a key exists
print("Name" in d)
print()

#11. Check if a value exists
print("Aman" in d.values())
print()

#12. Find the number of key-value pairs
print(len(d))
print()

#13. Copy the dictionary
d2 = d.copy()
print(d2)
print()

#14. Add a key only if it doesn't exist
d.setdefault("Country", "India")
print(d)
print()

#15. Iterate through all keys
for key in d.keys():
    print(key)
print()

#16. Iterate through all values
for value in d.values():
    print(value)
print()

#17. Iterate through all key-value pairs
for key, value in d.items():
    print(key, ":", value)
print()

#18. Clear the dictionary
d.clear()
print(d)
print()
```

NESTED DICTIONARIES ->>

```python
student = {
    "Name": "Aman",
    "Marks": {
        "Python": 95,
        "Math": 90
    }
}

print(student["Marks"]["Python"])

# Printing the key values using for loop

for key, value in student.items():
    print(key, value)


```

```python
square = {x: x*x for x in range(1, 6)}
print(square)
```

```python
# Counting the elemts

text = ["apple", "banana", "apple", "orange", "banana"]

count = {}

for fruit in text:
    count[fruit] = count.get(fruit, 0) + 1

print(count)
```

```python
# SORTING

#sorted - The sorted() function takes a collection of items and puts 
#them in order (usually smallest to largest)

#dict - The dict() function is a creator tool. It takes data that is structured 
#like pairs and converts it into an actual Python Dictionary ({}

# BY key
d = {"b": 2, "a": 1, "c": 3}
print(dict(sorted(d.items())))

# BY value
d1 = {"a": 2, "b": 1, "c": 3}
print(dict(sorted(d1.items(), key=lambda x: x[1])))
```

```python
# Convert Lists to Dictionary

keys = ["Name", "Age", "City"]
values = ["Aman", 20, "Pune"]

d = dict(zip(keys, values))
print(d)
print()
# Convert Dictionary to Lists

print(list(d.keys()))
print(list(d.values()))
print(list(d.items()))
```
