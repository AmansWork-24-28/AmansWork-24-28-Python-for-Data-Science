# Single Dimension

```python
import numpy as np

arr = np.array([1,2,5,4,6,7,8,9])

print(arr[0]) # single element
print(arr[0:3]) # multiple
```

```python
sliced = arr[1:5] # we think its a copy of array but no it is a view(original part)
print(sliced)

sliced[0] = 0 # this will change the original array
arr
```

# Fancy Indexing & Boolean Masking

```python
# Fancy Indexing (Select Multiple Elements)

arr = np.array([10, 20, 30, 40, 50])
idx = [0, 2, 4] 
print(arr[idx])
```

```python
# Boolean Masking (Filter Data)

arr = np.array([10, 20, 30, 40, 50])
mask = arr > 25  
print(arr[mask]) 
```
