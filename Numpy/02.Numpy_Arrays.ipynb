```python
import numpy as np
```

# 1. Creating NumPy Arrays

Every element in array must have same datatype

```python
arr1 = np.array([1,2,3,4,5]) # 1D
print(arr1)
print()
arr2 = np.array([[1,1,1],[2,2,2],[2,5,3]],) # 2D
print(arr2)
```

```python
# Arrays from scratch

print(np.zeros((3,2))) # 3x2 with all 0
print()   
print(np.ones((2,2)))  # 2x2 with all 1
print()    
print(np.full((3,3),5))  # 3x3 with all 5
print()    
print(np.eye(3)) # identity matrix of 3
print()    
print(np.linspace(1,10,5)) # divides 1 to 10 into 5 equal parts
print()
print(np.arange(1,5,1))   # 1 to 5 with gap of 1
```

# 2.Array Properties

```python
arr = np.array([[10, 20, 30], [40, 50, 60]])

print("Shape:", arr.shape) 
print("Size:", arr.size)
print("Dimensions:", arr.ndim)
print("Data type:", arr.dtype) 
```

# 3. Reshaping Array

```python
arr = np.array([[1,2,5],[4,5,8]])
print(arr)
print()
print(arr.shape)
print()
reshaped = arr.reshape(3,2)
print(reshaped)
```

```python
# Flattening array ( 2D to 1D)

FLAT = arr.flatten()
FLAT
```
