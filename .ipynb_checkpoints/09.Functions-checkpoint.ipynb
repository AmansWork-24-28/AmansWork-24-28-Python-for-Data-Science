# TYPES OF FUNCTIONS IN PYTHON

There are two types of functions in python:
• Built in functions (Already present in python)
• User defined functions (Defined by the user)
Examples of built in functions includes len(), print(), range() etc.
The func1() function we defined is an example of user defined function




```python
# Example of simple func

def simple():
    print("Hello welcome")

simple()
```

```python
# Function with arguments

def argfunc(a , b):
    print(a+b)

argfunc(1,2) 
#argfunc() # will give erroe
```

```python
# Function with default argument

def dfunc(name = "unknown"):
    #print("My name is "+name)
    print(f"My name is {name} .")
    
dfunc()    # use the default value 
dfunc("Aman") # use the passed value
```
