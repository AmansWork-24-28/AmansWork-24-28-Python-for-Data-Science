Primarily these are the following data types in Python:
1. Integers
2. Floating point numbers
3. Strings
4. Booleans
5. None

```python
a = 5 #integer
print(a)
```

```python
b = 4.5 #floating
```

```python
s = "aman" #string
print(s)
```

```python
c = False #boolean
print(bool(False)) # False
print(bool(None))  # False
print(bool(0))     # False
print(bool(""))    # False
print(bool([]))    # False
print(c)
```

```python
c1 = True 
print(bool("Hello")) # True (non-empty string)
print(bool(42))      # True (non-zero number)
print(bool([1, 2]))  # True (non-empty list)
```
