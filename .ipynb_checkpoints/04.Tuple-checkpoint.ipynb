A tuple is an immutable data type in python

```python
a = () # empty tuple
a = (1,) # tuple with only one element needs a comma
a = (1,7,2) # tuple with more than one element
```

```python
# 1. Mixed Data Types
my_tuple = (42, 3.14, "Python", True)

# 2. Indexing and Slicing
print(my_tuple[0])    # Outputs: 42
print(my_tuple[-1])   # Outputs: True
print(my_tuple[1:3])  # Outputs: (3.14, 'Python')

# 3. Immutability (This will throw an error if uncommented!)
# my_tuple[2] = "Jupyter" # TypeError

# 5. Duplicates are allowed
duplicate_tuple = (1, 2, 2, 3, 3, 3)
print(duplicate_tuple) # Outputs: (1, 2, 2, 3, 3, 3)
```

```python
# Initialize a new tuple with different numbers (Note the parentheses)
t1 = (14, 3, 9, 22, 5, 11)
print("Original tuple:", t1)

# 1. Sorting a tuple
# t1.sort() does not exist. Instead, use the built-in sorted() function, 
# which returns a list, and convert it back to a tuple.
t1_sorted = tuple(sorted(t1))
print("Sorted tuple:        ", t1_sorted)

# 2. Reversing a tuple
# t1.reverse() does not exist. Use slicing [::-1] or tuple(reversed(t1)).
t1_reversed = t1[::-1]
print("Reversed tuple:      ", t1_reversed)

# 3. Appending to a tuple
# t1.append(8) does not exist. Concatenate another tuple containing 8.
# Note the comma (8,) which tells Python it's a tuple, not just an integer in parentheses.
t1_appended = t1 + (8,)
print("After appending 8:   ", t1_appended)

# --- The original tuple 't1' remains untouched throughout ---
print("\n--- Working with original tuple for index operations ---")

# 4. Inserting into a tuple
# t1.insert(3, 8) does not exist. Slice the tuple and insert the new element.
t1_inserted = t1[:3] + (8,) + t1[3:]
print("After inserting 8 at index 3:", t1_inserted)

# 5. Popping from a tuple
# t1.pop(2) does not exist. Extract the value and reconstruct the tuple without it.
popped_value = t1[2]
t1_popped = t1[:2] + t1[3:]
print("After 'popping' index 2:     ", t1_popped, f" (Popped value: {popped_value})")

# 6. Removing a specific value from a tuple
# t1.remove(22) does not exist. Find the index and skip it, or convert to a list first.
# Here is the clean way by converting to a list and back:
temp_list = list(t1)
temp_list.remove(22)
t1_removed = tuple(temp_list)
print("After removing 22:           ", t1_removed)
```
