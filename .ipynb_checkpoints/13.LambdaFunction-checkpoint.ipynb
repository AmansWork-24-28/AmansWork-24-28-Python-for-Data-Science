# When to Use Lambda Functions?

Use Lambda Functions When:  
1.The function is short and simple.  
2.Used temporarily inside another function (e.g., map , filter ).  
3.Avoiding defining a full function with def .  

Avoid Lambda Functions When:  
1.The function is complex (use def for readability).  
2.Multiple operations/statements are needed.  


# Syntax

lambda arguments: expression

```python
cube = lambda x:x**3
print(cube(3))
```

# 2. Using Lambda Functions with map() , filter() , and reduce()


```python
# map() - Applies a function to each element of an iterable

nums = [1,2,5,4,3,6]
square = list(map(lambda x:x**2,nums))
print(square)

```

```python
# filter() - Filters elements based on a condition.

nums = [1,2,5,4,3,6]
filtered = list(filter(lambda x:x%2==0,nums))
print(filtered)
```

```python
# reduce() -Reduces an iterable to a single value (requires functools.reduce ).

from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
subtraction = reduce(lambda x, y: x - y, numbers)
addition = reduce(lambda x, y: x + y, numbers)

print(product) 
print(addition)
print(subtraction)
```

# 3. Lambda with Multiple Arguments

```python
nums = lambda x,y:x*y
print(nums(2,5))

ifelse = lambda x:"Even" if x%2==0 else "Odd"
a = int(input("Enter no. to be checked : "))
print(ifelse(a))
```

# 4. Lambda in Sorting Functions

```python
students = [("Alice", 85), ("gob", 78), ("Charlie", 92)]
students.sort(key=lambda student:student[1])
print(students)
```
