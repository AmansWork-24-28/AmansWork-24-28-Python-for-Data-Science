WHILE LOOP

Syntax:
while (condition): # The block keeps executing until the condition is true
      body

# Body of the loop
+ In while loops, the condition is checked first. If it evaluates to true, the body of the loop
+ is executed otherwise not!
+ If the loop is entered, the process of [condition check & execution] is continued until
  the condition becomes False.

```python
a = 0
while(a!=11):
    print(a)
    a=a+1
```

```python
# printing items from lsit using while loop

l = [1,2,3,5,4,6,8]
i=0
while i<len(l):
    print(l[i])
    i=i+1
```

FOR LOOP 

Syntax -
for condn :


```python
# Using range function

for i in range(1,11):
    print(i)
```

```python
# To iterate (using keyword in)
l=["aman","santosh","gotad"]
for i in l:
    print(i)
```

```python
# reverse prinitng

for i in range(10, 0, -1):
    print(i)
```

```python
# nested
for i in range(3):
    for j in range(3):
        print(i, j)
```
